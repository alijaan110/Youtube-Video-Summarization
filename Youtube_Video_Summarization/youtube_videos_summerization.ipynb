```python
# Import necessary libraries
import os
import pytube
import whisper
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Function to download YouTube video and extract audio
def download_audio_from_youtube(url):
    youtube = pytube.YouTube(url)
    video = youtube.streams.filter(only_audio=True).first()
    out_file = video.download(output_path=".")
    base, ext = os.path.splitext(out_file)
    audio_file = base + '.mp3'
    os.rename(out_file, audio_file)
    return audio_file
```

```python
# Function to transcribe audio using OpenAI Whisper
def transcribe_audio(audio_file):
    model = whisper.load_model("base")
    result = model.transcribe(audio_file)
    return result["text"]
```

```python
# Function to summarize text using Hugging Face Transformers
def summarize_text(text):
    model_name = "t5-base"
    tokenizer = T5Tokenizer.from_pretrained(model_name)
    model = T5ForConditionalGeneration.from_pretrained(model_name)

    input_ids = tokenizer.encode("summarize: " + text, return_tensors="pt", max_length=512, truncation=True)
    summary_ids = model.generate(input_ids, max_length=512, min_length=128, length_penalty=2.0, num_beams=4, early_stopping=True)

    summary = tokenizer.decode(summary_ids[0], skip_special_tokens=True)
    return summary
```

```python
# Main function to get summary from YouTube URL
def get_youtube_summary(url):
    audio_file = download_audio_from_youtube(url)
    transcript = transcribe_audio(audio_file)
    summary = summarize_text(transcript)
    os.remove(audio_file)  # Clean up the audio file after processing
    return summary
```

```python
# Enter the YouTube URL
youtube_url = input("Enter the YouTube URL: ")

```

```python
# Get the summary
summary = get_youtube_summary(youtube_url)
print("\nSummary:\n", summary)
```
